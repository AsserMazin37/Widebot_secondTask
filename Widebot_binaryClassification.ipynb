Let's import the required libraries

```
import numpy as np
import pandas as pd
```

```
Data = pd.read_csv('training.csv', delimiter=',')
data_target = Data['classLabel']
Data = Data.drop('classLabel',axis=1)
```

```
print(Data.shape)
```

```
validation= pd.read_csv('validation.csv', delimiter=',')
validation_target = validation['classLabel']
validation = validation.drop('classLabel',axis=1)
```

```
print(validation.shape)
```

```
print(Data.isnull().sum())
```

```
print(validation.isnull().sum())
```

After exploring the data, the feauture named 'variable18' contains too many missing data so we will drop it from both training and validation datasets

```
Data = Data.drop('variable18' , axis = 1)
validation = validation.drop('variable18', axis=1)
```

Now let's copy our data to two new variables to manipulate them

```
train_data = Data.copy()
validation_data = validation.copy()
```

After careful observing and analysing our data we will use mean and mode imputation to cover our missing data

```
# Mode Imputation
def occupy_missing_data(new_data,old_data):
    new_data["variable1"].fillna(old_data['variable1'].value_counts().idxmax(), inplace=True)
    new_data["variable4"].fillna(old_data['variable4'].value_counts().idxmax(), inplace=True)
    new_data["variable5"].fillna(old_data['variable5'].value_counts().idxmax(), inplace=True)
    new_data["variable6"].fillna(old_data['variable6'].value_counts().idxmax(), inplace=True)
    new_data["variable7"].fillna(old_data['variable7'].value_counts().idxmax(), inplace=True)
```

```
occupy_missing_data(train_data,Data)
occupy_missing_data(validation_data,validation)
```

```
# Mean Imputation
def replace_numbers_using_mean(new_data,old_data):
    new_data['variable2'].fillna(old_data['variable2'].mean(), inplace=True)
    new_data['variable14'].fillna(old_data['variable14'].mean(), inplace=True)
    new_data['variable17'].fillna(old_data['variable17'].mean(), inplace=True)
```

```
replace_numbers_using_mean(train_data,Data)
replace_numbers_using_mean(validation_data,validation)
```

Now let's Normalize our data using MinMaxScaler() Func from sklearn

```
def transform(Data):
    transform_features = pd.DataFrame(data = Data)
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler() # The default range is (0, 1)
    numerical_features = ['variable2', 'variable3', 'variable8', 'variable14', 'variable15',
                          'variable17', 'variable19']

    transform_minMax_features = pd.DataFrame(data = transform_features)

    transform_minMax_features[numerical_features] = scaler.fit_transform(
            transform_features[numerical_features])
    return transform_minMax_features
```

```
train_data_transformed = transform(train_data)
train_data_transformed.head(10)
```

```
validation_data_transformed = transform(validation_data)
validation_data_transformed.head(10)
```

Now let's encode our categorical feautures using the get_dummies() func from numpy

```
final_training_data = pd.get_dummies(train_data_transformed)
final_training_data.head()
```

```
final_validation_data = pd.get_dummies(validation_data_transformed)
final_validation_data.head()
```

Time for some model training!!

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_training_data, data_target,
                                                    test_size = 0.25, random_state = 42)

# See the results of splitting
print('Number of training examples: {}'.format(X_train.shape[0]))
print('Number of validation examples: {}'.format(X_test.shape[0]))
```

Apply LogisticRegression to our training Data

```
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```
print('Accuracy: {}'.format(regressor.score(X_test, y_test)))
```

Now let's do some grid searching to obtain best parameters and best score

```
from sklearn.model_selection import GridSearchCV
```

First let's initialize the grid values for the tol and max_iter params

```
tol = [0.01,0.001,0.0001]
max_iter = [100,150,200]

param_grid = dict(tol=tol, max_iter=max_iter)
```

Now let's fit our training data to GridSearch

```
grid_model = GridSearchCV(estimator=regressor, param_grid=param_grid, cv=5)

grid_model_results = grid_model.fit(final_training_data,data_target)

# Summary
best_score, best_params = grid_model_results.best_score_,grid_model_results.best_params_
print("Best: %f using %s" % (best_score, best_params))
```
